```python
# Making changes to data
```

```python
df['Total']=df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed'] 
```

```python
df.head(5)
```

```python
# dropping an column
df=df.drop(columns=['Total']);
df
```

```python
# another way of adding column
df['Total']=df.iloc[:,4:10].sum(axis=1)
```

```python
df
```

```python
# moving column 'Total' to left
cols=list(df.columns.values)
df=df[cols[0:4] + [cols[-1]] + cols[4:12]]
df.head(5)
```

```python
#saving our data (exporting into desired format)
df.to_csv('modified.csv') # modified pokemon_data.csv
```

```python
df.to_csv('modified.csv',index=False) # removing index column 
```

```python
# to excel
df.to_excel('modified.xlsx',index=False);
```

```python
# Filtering Data
```

```python
df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison')]
```

```python
# Conditional Changes
```

```python
df.loc[df['Type 1']=='Fire','Type 1']='Flamer'
df
```

```python
df.loc[df['Type 1']=='Flamer','Type 1']='Fire'
df
```

```python
df.loc[df['Total']>500,['Generation','Legendary']]=['Test1','Test2']
df
```
